#Recriando o dataset da aula:

```python
import pandas as pd

data = {
    'Peso': [100, 120, 180, 200, 150],
    'Cor': [2, 3, 8, 9, 7],
    'Classe': [0, 0, 1, 1, 1]
}

df = pd.DataFrame(data)

display(df)
```

# Aplicando algoritmo k-NNA:

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X = df[['Peso', 'Cor']].values
y = df['Classe'].values

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X, y)

new_item = np.array([[140, 6]])

predicted_class = knn.predict(new_item)

print(f"Previsão de classe da nova fruta (Peso: 140, Cor: 6): {predicted_class[0]}")
```

# Visualização:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plot_original_df = df.copy()
plot_original_df['Tipo'] = 'Frutas Existentes'

new_item_data = {'Peso': [new_item[0, 0]], 'Cor': [new_item[0, 1]], 'Classe': [predicted_class[0]], 'Tipo': ['Nova Fruta']}
plot_new_item_df = pd.DataFrame(new_item_data)

combined_plot_df = pd.concat([plot_original_df, plot_new_item_df], ignore_index=True)

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=combined_plot_df,
    x='Peso',
    y='Cor',
    hue='Classe',
    style='Tipo',
    markers={'Frutas Existentes': 'o', 'Nova Fruta': 'X'},
    s=150,
    palette='viridis',
    edgecolor='black'
)

plt.title('Visualizacao: Peso vs Cor com Nova Fruta')
plt.xlabel('Peso')
plt.ylabel('Cor')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Categorias')
plt.show()
```
